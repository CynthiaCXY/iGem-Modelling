# Deterministic Modelling for LacY

```python
from __future__ import division, print_function # Imports from __future__ since we're running Python 2
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
%matplotlib inline
from scipy.integrate import odeint
from scipy.integrate import ode
```

```python
k_smr = 0.23
k_sr = 15
k_2r = 50
k_Neg_2r = 1e-3
k_r = 960
k_Neg_r = 2.4
k_dr1 = 3e-7
k_Neg_dr1 = 12
k_dr2 = 3e-7
k_Neg_dr2 = 4.8e3
k_s0my = 0.01
k_s1my = 0.5
k_sy = 30
k_p = 0.12
k_Neg_p = 0.1
k_ft = 6e4
k_t = 0.92

lambda_mr = 0.462
lambda_r = 0.2
lambda_i2r2 = 0.2
lambda_r2 = 0.2
lambda_my = 0.462
lambda_y = 0.2
lambda_yiex = 0.2
```

```python
def f(y, t):
     mr = y[0]
     r = y[1]
     r2 = y[2]
     o = y[3]
     i = y[4]
     i2r2 = y[5]
     my = y[6]
     lac_Y = y[7]
     yiex = y[8]
     #iex = y[9]

     f0 = k_smr - lambda_mr*mr
     f1 = k_sr*mr - 2*k_2r*r*r + 2*k_Neg_2r*r2 - lambda_r*r
     f2 = k_2r*r*r - k_Neg_2r*r2 - k_r*r2*o + k_Neg_r*(2.08-o) - k_dr1*i*i*r2 + k_Neg_dr1*i2r2 - lambda_r2*r2
     f3 = -k_r*r2*o + k_Neg_r*(2.08-o) + k_dr2*i*i*(2.08-o) - k_Neg_dr2*i2r2*o
     f4 = -2*k_dr1*r2*i*i + 2*k_Neg_dr1*i2r2 - 2*k_dr2*(2.08-o)*i*i + 2*k_Neg_dr2*o*i2r2 + k_ft*yiex + k_t*(iex-i) + 2*lambda_i2r2*i2r2 + lambda_yiex*yiex
     f5 = k_dr1*i*i*r2 - k_Neg_dr1*i2r2 + k_dr2*i*i*(2.08-o) - k_Neg_dr2*i2r2*o - lambda_i2r2*i2r2
     f6 = k_s0my*(2.08-o) + k_s1my*o - lambda_my*my
     f7 = k_sy*my + (k_ft + k_Neg_p)*yiex - k_p*lac_Y*iex - lambda_y*lac_Y
     f8 = -(k_ft + k_Neg_p)*yiex + k_p*lac_Y*iex - lambda_yiex*yiex  
     #f9 = k_t*(i-iex) - k_p*lac_Y*iex + k_Neg_p*yiex 
     return [f0, f1, f2, f3, f4, f5, f6, f7, f8]
```

```python
mr_0 = 0
r_0 = 0
r2_0 = 0
o_0 = 2.08
i_0 = 0
i2r2_0 = 0
my_0 = 0
lac_Y_0 = 0
yiex_0 = 0
#iex_0 = 100000
y0 = [mr_0, r_0, r2_0, o_0, i_0, i2r2_0, my_0, lac_Y_0, yiex_0]     
t  = np.linspace(0, 60, 5000) 
```
